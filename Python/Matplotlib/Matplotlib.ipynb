```python
####https://www.youtube.com/watch?v=QyhqzaMiFxk&list=PLQVvvaa0QuDfefDfXb9Yf0la1fPDKluPF&index=7
import matplotlib.pyplot as plt
```

```python
## Line Graph

x=[1,2,3,5]
y=[3,8,1,9]

plt.plot(x,y)
plt.xlabel('xaxis')
plt.ylabel('yaxis')
plt.title('Line Graph')
```

```python
### Legends

x=[1,2,3,5]
y=[3,8,1,9]

x1=[2,3,4,6]
y1=[4,9,2,10]

plt.plot(x,y, label='First')
plt.plot(x1,y1, label = 'Second')
plt.xlabel('xaxis')
plt.ylabel('yaxis')
plt.title('Graph Adding legend')
plt.legend()
```

```python
### Bar charts 

x=[1,2,3,5]
y=[3,8,1,9]

x1=[2,3,4,6]
y1=[7,9,2,10]

plt.bar(x,y, label='First',color='red')
plt.bar(x1,y1, label = 'Second',color='green')
plt.xlabel('xaxis')
plt.ylabel('yaxis')
plt.title('Bar charts ')
plt.legend()
```

```python
### Histograms
x = [24.91,27.61,36.87,43.19,50.7,44.87,52.63,59.79,47.73,50.7,47.13,34.89,45.46,37.22,35.98,
     27.57,33.46,43.49,39.85,39.58,36.96,31.05,39.74,30.64,40.87,45.54,22.33,25.65,35.21,35.82,
     29.38,23.98,38.52,54.66,46.01,44.73,39.96,38.1,59.6,32,36.26,27.62,53.6,31.86,52,49.48,30.34,26.38]

bins = [20,25,30,35,40,45,50,55,60,70,80]

plt.hist(x,bins, histtype='bar',color='red')
plt.xlabel('xaxis')
plt.ylabel('yaxis')
plt.title('Histograms')
```

```python
## Scatter Plots

x=[1,2,3,5,6,7,8,4,7,8]
y=[3,8,1,9,7,8,6,4,3,6]

plt.scatter(x,y, color='red',
            marker='^'  ## defines the symbol
            ,s=100)  ## s defines size of marker
plt.xlabel('xaxis')
plt.ylabel('yaxis')
plt.title('Scatter Plots')
```

```python
## Stack Plots

days     = [1,2,3,4,5]
sleeping = [7,8,6,11,7]
eating   = [2,3,4,3,2]
working  = [7,8,7,2,2]
playing  = [8,5,7,8,13]

plt.stackplot(days,sleeping,eating,working,playing, colors=['m','c','r','k'])
plt.xlabel('xaxis')
plt.ylabel('yaxis')
plt.title('Stack Plots')
```

```python
## Pie charts

slices = [7,8,6,11]
activities = ['sleeping','eating','working','playing']

a=plt.pie(slices,
         labels = activities, 
		 colors=['m','c','r','b'],
		 startangle=90,  ## startangle starts at desired angle
		 shadow = True,
		 explode = (0,0.5,0.0,0), # pull the pie out of the circle 
		 autopct = '%1.1f%%')  ## give % of each pie
a = plt.title('Pie charts')
```
