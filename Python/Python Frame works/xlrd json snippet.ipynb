```python
import xlrd 
  
loc = (r"C:\Users\chaitanya.vadapalli\Desktop\Prologis_GC RFP_Bid Form_Global (2019.10.28).xlsx") 
  
wb = xlrd.open_workbook(loc) 
sheet = wb.sheet_by_index(2) 
l=list()
for i in range(sheet.ncols):
    l.insert(i,'"'+sheet.cell_value(10, i)+'"')
print("[")
for i in range(sheet.nrows-1):
    if(i<sheet.nrows):
        print("{")
    else:
        break
    for j in range(len(l)):
        if(j<len(l)-1):
            print(l[j]+":"+'"'+str(sheet.cell_value(i+11,j))+'"'+",")
        else:
            print(l[j]+":"+'"'+str(sheet.cell_value(i+11,j))+'"')
    if(i<sheet.nrows):
        print("},")
    else:
        print("}")
        break
print("]")  
```
